# Bank Account

```python
account_bal = 0

def deposit(amount):
  global account_bal
  account_bal += amount
  print(f"Deposited {amount}, Current balance is {account_bal}")

def withdraw(amount):
  global account_bal
  if amount > account_bal:
    print("insufficent balance")
  else:
    account_bal -= amount
    print(f"Withdrawn:{amount}, Available balance is {account_bal}")
```

```python
deposit(1000)
```

```python
withdraw(500)
```
