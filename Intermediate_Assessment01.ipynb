```python
highest_qty = {}
for sale in sales_data:
  item = sale['item']
  quantity = sale['quantity']

  if item in highest_qty:
    highest_qty[item] += quantity
  else:
    highest_qty[item] = quantity


highest_item = '0'
highest_quantity = 0

for item, qty in highest_qty.items():
    if qty > highest_quantity:
        highest_quantity = qty
        highest_item = item

print(f"Item with highest quantity sold: {highest_item}")
print(f"Total quantity sold: {highest_quantity}")
```
