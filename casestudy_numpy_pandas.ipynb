# Numpy

```python
import numpy as np
import pandas as pd

df = pd.read_csv("/content/auto-mpg_DSA.csv")
```

```python
mpg = df['mpg'].values
mean_mpg = np.mean(mpg)
median_mpg = np.median(mpg)
std_dev_mpg = np.std(mpg)
car_above_25 = np.sum(mpg>25)

print("Mean MPG:", mean_mpg)
print("Median MPG:", median_mpg)
print("Standard Deviation:", std_dev_mpg)
print("Cars with MPG > 25:", car_above_25)
```

```python
cyl_arr = df['cylinders'].to_numpy()
car_names = df['car name'].to_numpy()

filtered_names = car_names[cyl_arr > 6].tolist()
print(filtered_names)
```

```python
weight_array = df['weight'].to_numpy()
percentiles = np.percentile(weight_array, [25, 50, 75])
print("25th, 50th, 75th percentiles of weight:", percentiles)
```

```python
displacement_arr = df['displacement'].to_numpy()
origin_arr = df['origin'].to_numpy()

avg_disp_origin2 = np.mean(displacement_arr[origin_arr == 2])
print("Average displacement for European cars (origin=2):", avg_disp_origin2)
```

```python
sorted_indices = np.argsort(df['model year'].to_numpy())[::-1]
top5_cars = df['car name'].to_numpy()[sorted_indices][:5]

print("Top 5 newest cars:", top5_cars)
```

```python
filtered_cars = df[(df['model year'] == 75) & (df['weight'] <3000)]
filtered_cars_selected = filtered_cars[['car name', 'weight', 'mpg']]
print(filtered_cars_selected)
```

```python
print(df.isna().sum())
```

```python
df['power_to_weight_ratio'] = df['horsepower'] / df['weight']
print(df[['car name', 'horsepower', 'weight', 'power_to_weight_ratio']].head())
```

```python
mean_mpg_by_origin = df.groupby('origin')['mpg'].mean()
print(mean_mpg_by_origin)
```

```python
top10_mpg = df.sort_values(by='mpg', ascending=False).head(10)
print(top10_mpg[['car name', 'mpg']])
```

```python
def performance_score(row):
    return row['mpg'] * row['acceleration'] / row['weight']

df['performance_score'] = df.apply(performance_score, axis=1)
print(df[['car name', 'performance_score']].head())
```

```python
summary_by_year = df.groupby('model year')[['mpg', 'weight', 'horsepower']].mean()
print(summary_by_year)
```
